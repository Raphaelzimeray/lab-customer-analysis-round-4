Boxplot ou boite Ã  moustache permet de rÃ©sumer une variable de maniÃ¨re simple et visuel, d'identifier les valeurs extrÃªmes et de comprendre la rÃ©partition des observations.
La valeur centrale du graphique est la mÃ©diane (il existe autant de valeur supÃ©rieures qu'infÃ©rieures Ã  cette valeur dans l'Ã©chantillon).
Les bords du rectangle sont les quartiles (Pour le bord infÃ©rieur, un quart des observations ont des valeurs plus petites et trois quart ont des
valeurs plus grandes, le bord supÃ©rieur suit le mÃªme raisonnement).
Les extrÃ©mitÃ©s des moustaches sont calculÃ©es en utilisant 1.5 fois l'espace interquartile (la distance entre le 1er et le 3Ã¨me quartile).

```python
fig = plt.gcf()
fig.set_size_inches(20,4)
meanprops = {'marker':'o', 'markeredgecolor':'black',
            'markerfacecolor':'firebrick'}
sbn.boxplot(x=numerical['customer lifetime value'], color ='gold', orient='h', medianprops = {'color':"green"},showmeans=True,meanprops=meanprops)
plt.title('Distribution de la colonne customer lifetime value ',color='black',fontsize=18)
```

```python
fig = plt.gcf()
fig.set_size_inches(20,4)
meanprops = {'marker':'o', 'markeredgecolor':'black',
            'markerfacecolor':'firebrick'}

sbn.boxplot(x=numerical['income'], color ='gold', orient='h', medianprops = {'color':"green"},showmeans=True,meanprops=meanprops)
plt.title('Distribution de la colonne income ',color='black',fontsize=18)
```

Essayons de faire une boucle sur toutes les colonnes afin de construire chaque boxplot

```python
for serie in numerical:
    fig = plt.gcf()
    fig.set_size_inches(20,4)
    meanprops = {'marker':'o', 'markeredgecolor':'black',
                'markerfacecolor':'firebrick'}

    #plt.style.use('dark_background')
    sbn.boxplot(x=numerical[serie], color ='gold', orient='h', medianprops = {'color':"green"},showmeans=True,meanprops=meanprops)
    plt.title('Distribution de la colonne customer lifetime value ',color='black',fontsize=18)
```

Ce serait formidable si nous pouvions dessiner une boÃ®te Ã  moustaches oÃ¹ chaque colonne a son propre axe y. De cette faÃ§on, les colonnes avec des Ã©chelles plus grandes n'affecteront pas visuellement les autres.

```python
# Avec Pandas matplotlib

numerical.plot(
    kind='box',  # type de graphique
    subplots=True, # pour pouvoir mettre plusieurs boxplot sur le mÃªme graphique
    sharey=False, # ne pas partager l'axe des y
    figsize=(17, 5) # taille du graphique
)

# increase spacing between subplots
plt.subplots_adjust(wspace=0.7)
#plt.xticks(color='red')
plt.show()
```

```python
# Avec Pandas matplotlib
numerical.plot.kde(
    #kind='box',  # type de graphique
    subplots=True, # permet de mettre plusieurs graphique sur le mÃªme support
    sharex=False, # ne pas partager l'axe des x
    figsize=(10, 20) # taille du graphique
)

# increase spacing between subplots
plt.subplots_adjust(wspace=3)
#plt.xticks(color='red')
plt.show()
```

# For the numerical variables, check the multicollinearity between the features. Please note that we will use the column total_claim_amount later as the target variable.

```python
# On utilise la mÃ©thode corr() pour calculer s'il existe une corrÃ©lation entre chaque variables quantitatives
corrMatrix = numerical.corr()
fig = plt.gcf()
fig.set_size_inches(10,8)
sbn.heatmap(corrMatrix, annot=True, cmap="Greens")
plt.xticks(rotation=15,fontsize=12)
plt.yticks(fontsize=12)
plt.title('MATRICE DE CORRELATION R (PEARSON) - Numerical datas', fontsize=18, color='black')
#plt.savefig('Mission3_1-Matrice-R-Quanti.png')
plt.show()
```
